```python
import pandas as pd
import os
from pathlib import Path
print(os.getcwd())
```

```python
# finding the path of all csv files
DATA_DIR = Path(r"D:\Learning Data Science\Project 1_Web Scraping\Seasons")
```

```python
season = {
    "Premier24": "2023-24",
    "Premier25": "2024-25",
    "Premier26": "2025-26"
}

leagues = ["E0", "E1", "E2", "E3", "EC"]
frames = []

for folder_name, season_label in season.items():
    
    for league in leagues:
        file_path = DATA_DIR / folder_name / f"{league}.csv"

        if not file_path.exists():
            print("Missing:", file_path)
            continue

        print("Loading:", file_path)
        df = pd.read_csv(file_path, encoding="unicode_escape")

        df["Season"] = season_label
        df["League"] = league

        frames.append(df)

# Combine everything
if not frames:
    raise ValueError("No CSV files were loaded — check folder names")

df_all = pd.concat(frames, ignore_index=True)
print("Final shape:", df_all.shape)
 
    
```

3 Organize All The Data in a Dictionary

```python
# create a dictionary with original name of league as key
dict_leagues = {
    'English Premier League':'E0',
    'English League 1':'E1',
    'English League 2':'E2',
    'English League 3':'E3',
    'European Competitions':'EC'
}
```

```python
# get a dictionary element 
dict_leagues['English Premier League']
```

```python
import os

# Folder containing season folders
base_path = r"D:\Learning Data Science\Project 1_Web Scraping\Seasons"

dict_historical_data = {}

# Map folder names to seasons
season_folders = {
    23: "Premier24",
    24: "Premier25",
    25: "Premier26"
}

# Looping through key elements
for league_name, league_code in dict_leagues.items():
    frames = [] # one dataframe per league
    for season, folder in season_folders.items():
        file_path = os.path.join(base_path, folder, f"{league_code}.csv")
        df = pd.read_csv(file_path, encoding="unicode_escape")
        df.insert(1, 'Season', season)
        frames.append(df)
    
    # Combine all seasons for this league
    dict_historical_data[league_name] = pd.concat(frames, ignore_index=True)
```
